```python
from langchain_text_splitters import CharacterTextSplitter

#CharacterTextSplitter를 사용하여 텍스트를 청크(chunk)로 분활하는 코드
text_splitter = CharacterTextSplitter(
    # 텍스트를 분활 할 때 사용할 구분자를 지정합니다. 기본값은 "\n\n"입니다.
    separator = '\n\n',
    # 분활된 텍스트 청크의 최대 크기를 지정합니다.(문자 수)
    chunk_size = 210,
    #분활된 텍스트 청크 간의 중복되는 문자 수를 지정합니다.
    chunk_overlap  = 0,
    # 텍스트의 길이를 계산하는 함수를 지정합니다.
    length_function = len,
)
```

```python
# 텍스트를 청크로 분활합니다.
texts = text_splitter.create_documents([file])
print(len(texts[0].page_content)) #분활된 문서의 개수를 출력합니다.
print(texts[0]) # 분활된 문서 중 첫 번째 문서를 출력합니다.
```

```python
metadatas = [
    {"document": 1},
    {"document": 2},
]
    # 문서에 대한 메타데이터 리스트를 정의합니다.
documents = text_splitter.create_documents(
      [
          file,
          file,
      ], #분활할 텍스트 데이터를 리스트로 전달합니다.
      metadatas=metadatas, #각 문서에 해당ㄷ하는 메타 데이터를 전달합니다.
  )
print(documents[0]) # 분활된 문서 중 첫 번째 문서를 출력합니다.
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

# 파일 경로 설정
file_1_path = "/content/data/appendix-keywords.txt"
file_2_path = "/content/data/자기소개서.txt"

# 파일 읽기
with open(file_1_path, "r", encoding="utf-8") as f1:
    file_1 = f1.read()

with open(file_2_path, "r", encoding="utf-8") as f2:
    file_2 = f2.read()

# 텍스트 분할기를 설정합니다 (필요에 따라 max_chunk_size를 조정하세요).
text_splitter = CharacterTextSplitter(
    # 텍스트를 분활 할 때 사용할 구분자를 지정합니다. 기본값은 "\n\n"입니다.
    separator = '\n\n',
    # 분활된 텍스트 청크의 최대 크기를 지정합니다.(문자 수)
    chunk_size = 210,
    #분활된 텍스트 청크 간의 중복되는 문자 수를 지정합니다.
    chunk_overlap  = 0,
    # 텍스트의 길이를 계산하는 함수를 지정합니다.
    length_function = len,
)

# 문서에 대한 메타데이터 리스트를 정의합니다.
metadatas = [
    {"document": 1},
    {"document": 2},
]

# 분할된 문서를 생성합니다.
documents = text_splitter.create_documents(
    [file_1, file_2],  # 분할할 텍스트 데이터를 리스트로 전달합니다.
    metadatas=metadatas  # 각 문서에 해당하는 메타데이터를 전달합니다.
)

# 분할된 문서 중 첫 번째 문서를 출력합니다.
print(documents[32])
```
